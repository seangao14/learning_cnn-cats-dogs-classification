```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style

style.use("ggplot")
header_names = ["MODEL_NAME", "TIME", "TRAIN_ACC", "TRAIN_LOSS", "VAL_ACC", "VAL_LOSS"]

# model-1599954702        ????
# model-1599955615        shuffled  10? epochs
# model-1599955848        not shuffled 10? epochs
# model-1599956524        shuffled 10 epochs
# model-1599993837        softmax instead of log_softmax
# model-1599994053        next 4 epochs of softmax
```

```python
def create_graph(model_name):
    content = pd.read_csv("model.log", names=header_names)
    content = content.loc[content["MODEL_NAME"]==model_name]
    
    ax1 = plt.subplot2grid((2,1), (0,0))
    ax2 = plt.subplot2grid((2,1), (1,0))
    
    ax1.plot(content["TIME"], content["TRAIN_ACC"], label="train acc", linewidth=1)
    ax1.plot(content["TIME"], content["VAL_ACC"], label="val acc", linewidth=1)
    ax1.legend(loc=2)
    
    ax2.plot(content["TIME"], content["TRAIN_LOSS"], label="train loss", linewidth=1)
    ax2.plot(content["TIME"], content["VAL_LOSS"], label="val loss", linewidth=1)
    ax2.legend(loc=2)
    
    plt.show()
```

```python
create_graph(model_name="model-1600204299")
```
